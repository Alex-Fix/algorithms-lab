```python
viz = AlgoVisualizer()
data = [64, 25, 12, 22, 11]

# RECORDING PHASE
def selection_sort_recorded(arr):
    n = len(arr)
    viz.record_step(arr, title="Initial State") # Record start
    
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        # Record before swap (highlighting the two being compared/swapped)
        viz.record_step(arr, highlight_indices=[i, min_idx], title=f"Finding min for index {i}")
        
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        
        # Record after swap
        viz.record_step(arr, title=f"Swapped {arr[i]} and {arr[min_idx]}")

selection_sort_recorded(data)

# INTERACTIVE PHASE
viz.render()
```
