```python
import sys, os
sys.path.append(os.path.abspath('../../'))
from shared_py.visualizer import AlgoVisualizer
```

```python
viz = AlgoVisualizer()
data = [64, 25, 12, 22, 11]

# RECORDING PHASE
def bubble_sort_recorded(arr):
    n = len(arr)
    viz.record_step(arr, title="Initial State") # Record start
    
    while True:
        swapped = False

        for i in range(n - 1):
            j = i + 1
            if arr[i] > arr[j]:
                # Record before swap (highlighting the two being compared/swapped)
                viz.record_step(arr, highlight_indices=[i, j], title=f"Finding min for index {i}")
                
                arr[i], arr[j] = arr[j], arr[i]

                # Record after swap
                viz.record_step(arr, title=f"Swapped {arr[i]} and {arr[j]}")
                
                swapped = True

        if not swapped:
            break


bubble_sort_recorded(data)

# INTERACTIVE PHASE
viz.render()
```
